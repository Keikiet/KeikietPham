### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

1) Pyplot approach
2) Object-oriented approach (recommended). Here you create a objects like figure and axes and then call methods on them to create a and customize the plot. This approach also synergizes better with Numpy.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

1) Figure: A figure is the entire window or page where your plots and visualizations are drawn. It acts as a container that holds all the elements of a plot, such as axes, titles, labels, etc. You can think of it as the top-level container for everything that you see in a plot.

2) Axes: Axes are individual plots or graphs within the figure. Each figure can have multiple axes, arranged in a grid. Axes are where data is actually plotted. An axes object includes the x-axis and y-axis, along with labels, ticks, and other elements.

3) Axis: Axes are the actual number-line-like objects. In Matplotlib, an axis refers to the x-axis or y-axis of a plot. It defines the range of values displayed along an axis and provides ticks and labels to help interpret the data.

4) Artist: In Matplotlib, nearly everything you see on a figure is an artist. This includes the Figure, Axes, Axis, text objects, line objects, and more. Artists are the basic building blocks used to create a plot. They can be manipulated to customize the appearance of the plot.

numpy.array

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Generates 20 values ranging from -4 to 4
x = np.linspace(-4, 4, 20)

# Calculate y values
y = x**2 

# Create figure and axes
fig, (ax) = plt.subplots()

# Plot the function
ax.plot(x, y, label = "x^2")
ax.plot(x, y+2, label = "x^2+2")

# Set title, labels and legend
ax.set_title("plot of y=x^2 - Object-based approach")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

plt.show()
```

```python
#State-based approach

# Generate x values from -4 to 4
x = np.linspace(-4, 4, 100)

# Calculate y values
y = x**2

# Plot the function
plt.plot(x, y)

# Set title and labels
plt.title("Plot of y = x^2 - pyplot approach")
plt.xlabel("x")
plt.ylabel("y")

# Show the plot
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha = .4, color='black')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Bar plot
ax2.bar(names, values, color='black')
ax2.set_title('Bar Plot')
ax2.set_xlabel('Fruits')
ax2.set_ylabel('Quantity')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Example codes:

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```
